```python
import numpy as np
import pandas as pd
import ast
import re
```

```python
def fetch_name(obj): 
    L=[]
    if obj is not (None or []):
        for i in ast.literal_eval(obj):
            L.append(i['name']);
        return L
    else:
        return []
```

```python
def fetch_name2 (obj):
    try:
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name']);
        return L
    except:
        return []
```

```python
def fetch_names(obj): 
    
    if isinstance(obj, str) and '{' in obj:
        # print(obj)
        dic = ast.literal_eval(obj)

        return dic['name']
    
```

```python
#Creo funcion para calcular la columa retorno
def calculate_return2(row):
    revenue = row['revenue']
    budget = row['budget']
    if pd.isna(revenue) or pd.isna(budget):
        return 0
    elif budget.isnumeric() and float(budget) > 0:
        return revenue / float(budget)
    else:
        return 0
```

```python
#lleno los vacios y nulos con cero
am_df['revenue'] = am_df['revenue'].fillna(0)
am_df['budget'] = am_df['budget'].fillna(0)
```

```python
#Aplico la funcio apply para llenar la nueva columna
am_df['return'] = am_df.apply(calculate_return2, axis=1)
```

```python
# Definir la función para extraer el nombre del director
def extract_director_name(row):
    crew_list = eval(row['crew'])  # Convertir el string en una lista de diccionarios
    director_name = None
    for crew_member in crew_list:
        if crew_member['job'] == 'Director':
            director_name = crew_member['name']
            break
    return director_name
```

```python
autores_df['director_name'] = autores_df.apply(extract_director_name, axis=1)
```
